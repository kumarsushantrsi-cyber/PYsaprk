### Reading Data & Inspecting the Schema
Let's read our first dataset (`sales.csv`).
In Spark, data is loaded into a **DataFrame**, which is similar to a table in a relational database.

*Note: Spark uses "Lazy Evaluation". Reading the file doesn't actually load all the data into memory until you trigger an **Action** (like `.show()` or `.count()`).*


```python
# Read the CSV file into a DataFrame
# header=True tells Spark the first row contains column names
# inferSchema=True tells Spark to guess the data types (integer, string, etc.)
sales_df = spark.read.csv("sales.csv", header=True, inferSchema=True)

# Action 1: Print the schema (column names and data types)
print("--- Schema ---")
sales_df.printSchema()

# Action 2: Show the first 5 rows
print("--- Data Preview ---")
sales_df.show(5)

# Action 3: Count total rows
total_rows = sales_df.count()
print(f"Total rows in sales dataset: {total_rows}")
```

### Transformations vs. Actions
- **Transformations** (e.g., `select`, `filter`): These create a *new* DataFrame from an existing one. They are lazy and don't execute immediately. Spark just adds them to an execution plan.
- **Actions** (e.g., `show`, `count`, `collect`): These trigger the actual execution of the plan across the cluster to return a final result.


```python
# 1. SELECT: Choose specific columns
selected_df = sales_df.select("order_id", "product_id", "quantity")
print("--- Selected Columns ---")
selected_df.show(3)

# 2. FILTER: Keep only rows that match a condition
# Let's find orders where the quantity is greater than 2
high_qty_df = sales_df.filter(sales_df["quantity"] > 2)

print("--- Filtered Data (Quantity > 2) ---")
high_qty_df.show(5)
```
